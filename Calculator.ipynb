```python
import numpy as np
import matplotlib.pyplot as plt

def triangle_area(): # This is basically a line where it difines the function
    b = float(input("Enter the base of the triangle: "))  # lets the user input the base
    h = float(input("Enter the height of the triangle: "))# lets the user input the height
    area = 0.5 * b * h # the formula for ara of triangke
    print(f"The area of the triangle is: {area}") # prints the answer in a string

def rectangle_area(): #same as last one but just area of rectangle now
    b = float(input("Enter the length of the rectangle: "))
    h = float(input("Enter the width of the rectangle: "))
    area = b * h
    print(f"The area of the rectangle is: {area}")

def square_area(): #area of the square
    b = float(input("Enter the side of the square: "))
    area = b ** 2
    print(f"The area of the square is: {area}")

def circle_area(): #area of the cricle
    b = float(input("Enter the radius of the circle: "))
    pi = np.pi
    area = pi * (b ** 2)
    print(f"The area of the circle is: {area}")

def addition():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    print(f"The sum is: {num1 + num2}")

def subtraction():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    print(f"The difference is: {num1 - num2}")

def multiplication():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    print(f"The product is: {num1 * num2}")

def division():
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    if num2 != 0:
        print(f"The quotient is: {num1 / num2}")
    else:
        print("Division by zero not possible try again.")

def plot_trigonometric_graph():
    choice = input('Choose from sin, cos or tan: ') #let's the user choice which one

    if choice == 'sin':
        x = np.linspace(-10, 10, 400) #here on the x axis the numbers start from -4 and ends at 4 and it used to plot on the graph.
        y = np.sin(x)  # this line allows us to fnd  the y values for the sin grpah by plugging in the x values.
        plt.plot(x, y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.grid(True)
        plt.show()
    elif choice == 'cos':
        x = np.linspace(-10, 10, 400)
        y = np.cos(x)
        plt.plot(x, y)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.grid(True)
        plt.show()
    elif choice == 'tan':
        x = np.linspace(-10, 10, 400)
        y = np.tan(x)
        plt.plot(x, y)
        plt.ylim(-6, 6) # This is a fucntion which sets sets the y axis of the plot from only -6 to 6. This allows us to understand the graph better. reffered this to stack over flow.
        plt.xlabel('x')
        plt.ylabel('y')
        plt.grid(True)
        plt.show()

    else:
        print("wrong choice, try again") #if the user input's anything else then this line will them to input from the option again

def plot_polynomial_graph():
    equation = input("Enter a polynomial equation using 'x' as the variable: ")
    x = np.linspace(-10, 10, 400) #This line basically tells us that the range values will be from -10 to 10 which is basically our x values. It creates 400 evenely spaced vaules between -10 and 10.

    if 'x' in equation:
        y = eval(equation) #I had to do research for this so i went to https://www.w3schools.com/python/ref_func_eval.asp. This is where I understood how to use it
     #Basically if x is in the equation then this line will use the eval fucntion to evulate it which is basically evaluating the string.

        plt.plot(x, y)
        plt.title(equation)  # this line just prints the title of the graph which is the equation entered by the user.
        plt.xlabel('x', color='green')
        plt.ylabel('y', color='red')
        plt.grid(True)

        user_choice = input("Do you want a linear line? (yes/no) ") #aksking the user if they also wan't the linear line on the same graph

        if user_choice.lower() == "yes":
            linear_equation = input("Give me an example of a linear equation: ")
            y2 = eval(linear_equation) #evaultes the equation again
            plt.plot(x, y2)
            intersection_x = np.intersect1d(x, x[np.isclose(y, y2)]) # for the poi I used ai because I could not figure out how to do it. I understand that
            #It find's the x vaules where two lines intersect, bascially by comparing the y vaules of the two lines and then choosing the x values that are close enough to the y.
            # The function intersect1d just basically find's the x vaules for the intersection and the np,iscole just check's if they are close enough or not.
            intersection_y = np.intersect1d(y, y2) # this just basically find the common y vaule between y and y2 which absiclaly tells us the y vaule of the intersection points.

        if len(intersection_x) > 0 and len(intersection_y) > 0: # this line just checks if there are any intersetcion points on the x and y axis. If there is then it prin't s the next line
            print(f"Point of intersection: ({intersection_x[0]}, {intersection_y[0]})") # print the cordinates of the intersection
            plt.plot(intersection_x, intersection_y, 'ro', ) # This plots the inetrsetcion with the colour red
        else:
            print("No poi")
        plt.show()

def math_calculator():
    print("Choose Option 1 or 2:")
    print("1. Area of rectangle, triangle, circle, Square and addition, subtraction, Multiplication and division ")
    print("2. Graph  ")
    choice = input("Enter your choice (1/2): ")

    if choice == "1":
        print("1. Area triangle")
        print("2. Area rectangle")
        print("3. Area square")
        print("4. Area circle")
        print("5. Addition")
        print("6. Subtraction")
        print("7. Multiplication")
        print("8. Division")
        Math_choice = input("Enter your choice (1,2,3,4,5,6,7,8): ")
        if Math_choice == "1":
            triangle_area()
        elif Math_choice == "2":
            rectangle_area()
        elif Math_choice == "3":
            square_area()
        elif Math_choice == "4":
            circle_area()
        elif Math_choice == "5":
            addition()
        elif Math_choice == "6":
            subtraction()
        elif Math_choice == "7":
            multiplication()
        elif Math_choice == "8":
            division()
        else:
            print("Invalid choice.")

    elif choice == "2":
        print("1. Trigonometric Graph")
        print("2. Polynomial Graph")
        graph_choice = input("Enter your choice (1/2): ")

        if graph_choice == "1":
            plot_trigonometric_graph()
        elif graph_choice == "2":
            plot_polynomial_graph()
        else:
            print("Invalid choice.")

    else:
        print("Invalid choice.")

math_calculator()
```
